```python
#using convert dtypes I am able to change the coulmn type to something that is usable.
df = df.convert_dtypes()
```

```python
#view the new coulmn types
df.dtypes
```

```python
df.count()
```

```python
#remove the commas and dollar signs in Price and Odometer and convert to integer columns
df['Price'] = df['Price'].str.replace(',', '')
df['Price'] = df['Price'].str.replace('$', '')
df['Odometer'] = df['Odometer'].str.replace(',','')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Price"] = pd.to_numeric(df['Price'])
df['Odometer'] = pd.to_numeric(df['Odometer'])
```

```python
#create dataset of just nissan vehicle sales
df_nissans = df[df['Make'] == 'NISSAN']
```
